```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/lasyasrikantamneni/Downloads/sonyNetworkCamera_cleaned.csv') 

df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Colours for the boxplots
box_color = 'lightblue' 
whisker_color = 'gray'  
median_color = 'red'     
cap_color = 'purple'  

# Boxplot with colours allocatted
boxplot = df_numeric.boxplot(figsize=(10, 10), vert=False,
                                boxprops=dict(color=box_color),
                                whiskerprops=dict(color=whisker_color),
                                capprops=dict(color=cap_color),
                                medianprops=dict(color=median_color))

plt.title('Box Plot of Numeric Columns')
plt.show()
```
