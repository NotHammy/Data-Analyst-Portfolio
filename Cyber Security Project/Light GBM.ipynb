```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
from sklearn.neural_network import MLPClassifier
import joblib
from imblearn.over_sampling import SMOTE
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from google.colab import files
from google.colab import drive
from sklearn.preprocessing import StandardScaler

drive.mount('/content/drive')
df_july = pd.read_csv("/content/drive/MyDrive/july_paper_features_unique.csv")

X_train = df_july.drop('device', axis=1)
y_train = df_july['device']

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

## Experiment on hyperparameters
# param_grid = {
#     'hidden_layer_sizes': [(50,), (100,)],
#     'learning_rate': ['adaptive'],
#     'learning_rate_init': [0.001, 0.01, 0.1],
#     'alpha': [0.0001, 0.001, 0.01],
#     'solver': ['adam'],
# }

# grid_search = GridSearchCV(MLPClassifier(max_iter=200, random_state=42), param_grid, cv=3, scoring='f1_weighted')
# grid_search.fit(X_train_pca, y_train_resampled)

# print("Best Parameters:", grid_search.best_params_)
# best_model = grid_search.best_estimator_

mlp = MLPClassifier(hidden_layer_sizes=(50,), max_iter=500, random_state=42,
                    learning_rate='adaptive',
                    learning_rate_init=0.001, alpha=0.001,
                    solver='adam')
mlp.fit(X_train_scaled, y_train)
```
