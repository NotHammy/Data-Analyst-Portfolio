```python
import pandas as pd
import ipaddress
```

```python
# Dictionary of data types for every column
dtype_dict = {
    # Identifiers for Source and Destination
    'protocolIdentifier': 'float64',
    'protocolName': 'string',
    'sourceMacAddress': 'object',
    'sourceIPv4Address': 'object',
    'sourceTransportPort': 'float64',
    'destinationMacAddress': 'object',
    'destinationIPv4Address': 'object',
    'destinationTransportPort': 'float64',
    'serviceType': 'string',
    # General flow information
    'flowStartMilliseconds': 'string',
    'flowEndMilliseconds': 'string',
    'flowDurationMilliseconds': 'float64',
    'flowAttributes': 'string', #'float64',
    'flowEndReason': 'string',
    'packetTotalCount': 'float64',
    'octetTotalCount': 'float64',
    'reverseFlowAttributes': 'float64',
    'reverseFlowDeltaMilliseconds': 'float64',
    'reversePacketTotalCount': 'float64',
    'reverseOctetTotalCount': 'float64',
    # TCP Flags
    'initialTCPFlags': 'object', 
    'unionTCPFlags': 'object',
    'reverseInitialTCPFlags': 'object',
    # Packet data
    'smallPacketCount': 'float64',
    'largePacketCount': 'float64',
    'nonEmptyPacketCount': 'float64',
    'firstNonEmptyPacketSize': 'float64',
    'maxPacketSize': 'float64',
    'dataByteCount': 'float64',
    'bytesPerPacket': 'float64',
    'averageInterarrivalTime': 'float64',
    'standardDeviationInterarrivalTime': 'float64',
    # Reverse packet data
    'reverseSmallPacketCount': 'float64',
    'reverseLargePacketCount': 'float64',
    'reverseNonEmptyPacketCount': 'float64',
    'reverseFirstNonEmptyPacketSize': 'float64',
    'reverseMaxPacketSize': 'float64',
    'reverseDataByteCount': 'float64',
    'reverseBytesPerPacket': 'float64',
    'reverseAverageInterarrivalTime': 'float64',
    'reverseStandardDeviationInterarrivalTime': 'float64',
}

# Read the CSV file using the dictionary
######################################## (1) USER TO CHANGE PATH TO CSV FILE BELOW ########################################### 
path_to_csv_file = 'mapped_planex_camera_one_shot!.csv'
######################################## (1) USER TO CHANGE PATH TO CSV FILE ABOVE ########################################### 

df = pd.read_csv(path_to_csv_file, dtype=dtype_dict, parse_dates=['flowStartMilliseconds', 'flowEndMilliseconds'], 
                 usecols=['protocolIdentifier',
                        'protocolName',
                        'sourceMacAddress',
                        'sourceIPv4Address',
                        'sourceTransportPort',
                        'destinationMacAddress',
                        'destinationIPv4Address',
                        'destinationTransportPort',
                        'serviceType',
                        'flowStartMilliseconds',
                        'flowEndMilliseconds',
                        'flowDurationMilliseconds',
                        'flowAttributes',
                        'flowEndReason',
                        'packetTotalCount',
                        'octetTotalCount',
                        'reverseFlowAttributes',
                        'reverseFlowDeltaMilliseconds',
                        'reversePacketTotalCount',
                        'reverseOctetTotalCount',
                        'initialTCPFlags', 
                        'unionTCPFlags',
                        'reverseInitialTCPFlags',
                        'smallPacketCount',
                        'largePacketCount',
                        'nonEmptyPacketCount',
                        'firstNonEmptyPacketSize',
                        'maxPacketSize',
                        'dataByteCount',
                        'bytesPerPacket',
                        'averageInterarrivalTime',
                        'standardDeviationInterarrivalTime',
                        'reverseSmallPacketCount',
                        'reverseLargePacketCount',
                        'reverseNonEmptyPacketCount',
                        'reverseFirstNonEmptyPacketSize',
                        'reverseMaxPacketSize',
                        'reverseDataByteCount',
                        'reverseBytesPerPacket',
                        'reverseAverageInterarrivalTime',
                        'reverseStandardDeviationInterarrivalTime'])

# Reorder the columns of the dataframe
df = df[['protocolIdentifier',
        'protocolName',
        'sourceMacAddress',
        'sourceIPv4Address',
        'sourceTransportPort',
        'destinationMacAddress',
        'destinationIPv4Address',
        'destinationTransportPort',
        'serviceType',
        'flowStartMilliseconds',
        'flowEndMilliseconds',
        'flowDurationMilliseconds',
        'initialTCPFlags', 
        'unionTCPFlags',
        'reverseInitialTCPFlags',
        'flowAttributes',
        'flowEndReason',
        'packetTotalCount',
        'octetTotalCount',
        'reverseFlowAttributes',
        'reverseFlowDeltaMilliseconds',
        'reversePacketTotalCount',
        'reverseOctetTotalCount', 
        'smallPacketCount',
        'largePacketCount',
        'nonEmptyPacketCount',
        'firstNonEmptyPacketSize',
        'maxPacketSize',
        'dataByteCount',
        'bytesPerPacket',
        'averageInterarrivalTime',
        'standardDeviationInterarrivalTime',
        'reverseSmallPacketCount',
        'reverseLargePacketCount',
        'reverseNonEmptyPacketCount',
        'reverseFirstNonEmptyPacketSize',
        'reverseMaxPacketSize',
        'reverseDataByteCount',
        'reverseBytesPerPacket',
        'reverseAverageInterarrivalTime',
        'reverseStandardDeviationInterarrivalTime']]

# Removable | Displays the DataFrame
print(f"{df.dtypes}\n")
df.head(10)
```

```python
# Ensure flowAttributes is a float
df['flowAttributes'] = df['flowAttributes'].astype('float')

# List of Reverse Columns data
reverse_columns = [
    'reverseSmallPacketCount',
    'reverseLargePacketCount',
    'reverseNonEmptyPacketCount',
    'reverseFirstNonEmptyPacketSize',
    'reverseMaxPacketSize',
    'reverseDataByteCount',
    'reverseBytesPerPacket',
    'reverseAverageInterarrivalTime',
    'reverseStandardDeviationInterarrivalTime'
]

invalid = pd.to_numeric(df['reversePacketTotalCount'], errors='coerce').isna()
df = df[~invalid]
df['reversePacketTotalCount'] = df['reversePacketTotalCount'].astype('float')

for col in reverse_columns:
    # Fill in missing values in the Reverse Columns
    df.loc[df['reversePacketTotalCount'] == 0, col] = df.loc[df['reversePacketTotalCount'] == 0, col].fillna('0')
```

```python
df['flowEndReason'] = df['flowEndReason'].fillna('none')
df['flowStartMilliseconds'] = pd.to_datetime(df['flowStartMilliseconds'], errors='coerce')
df['flowEndMilliseconds'] = pd.to_datetime(df['flowEndMilliseconds'], errors='coerce')
```

```python
# Check if IP address is IPv4 or else return False
def check_IP(address):
    try:
        return type(ipaddress.ip_address(address)) == ipaddress.IPv4Address
    except ValueError:
        return False

# Keep only IPv4 Addresses
df = df[df['sourceIPv4Address'].apply(check_IP)]
df = df[df['destinationIPv4Address'].apply(check_IP)]

# Extract date and time into separate columns
df['flowStartMillisecondsDate'] = df['flowStartMilliseconds'].dt.date
df['flowEndMillisecondsDate'] = df['flowEndMilliseconds'].dt.date

df['flowStartMillisecondsTime'] = df['flowStartMilliseconds'].dt.time
df['flowEndMillisecondsTime'] = df['flowEndMilliseconds'].dt.time

# Drop flowStartMilliseconds and flowEndMilliseconds                                                
df.drop(['flowStartMilliseconds', 'flowEndMilliseconds'], inplace=True, axis=1)

# Adding in new columns

######################################## (2) USER TO CHANGE VALUES BELOW THIS LINE ########################################### 
# Check the KDDI IoT Github for the names, classes and Mac Address
df['deviceName'] = 'PLANEX camera one shot!'
df['deviceClass'] = 'Network camera'
df['IoTMacAddress'] = '00:1b:c7:fa:c3:e6'
######################################## (2) USER TO CHANGE VALUES ABOVE THIS LINE ########################################### 

# Trim MacAddress columns (there's an extra colon at the end of all the values)
df['sourceMacAddress'] = df['sourceMacAddress'].str.rstrip(':')
df['destinationMacAddress'] = df['destinationMacAddress'].str.rstrip(':')

# Keep only the records where the IoT Device was the Source
df = df[df['sourceMacAddress'] == df['IoTMacAddress']]

# Renaming the Source and Destination columns to IoT and Server
df.rename(columns={'sourceIPv4Address': 'IoTIPAddress', 'sourceTransportPort': 'IoTTransportPort', 
                   'destinationIPv4Address': 'serverIPAddress', 'destinationMacAddress': 'serverMacAddress', 'destinationTransportPort': 'serverTransportPort'}, inplace=True)

# Drop the Source Mac Address which is now just the same as the IoT Mac Address
df = df.drop(columns='sourceMacAddress')
```

```python
# Drop duplicate rows
df.drop_duplicates(inplace=True)

# Drop the corrupted or remaining records with missing values
df = df.dropna()

# Reset the indices of the records
df = df.reset_index(drop=True)

# Reorder the columns of the dataframe
df = df[['deviceName',
         'deviceClass',     
        'IoTMacAddress',
        'IoTIPAddress',
        'IoTTransportPort',
        'serverMacAddress',
        'serverIPAddress',
        'serverTransportPort',
        'serviceType',
        'protocolIdentifier',
        'protocolName',
        'flowStartMillisecondsDate',
        'flowEndMillisecondsDate',
        'flowStartMillisecondsTime',
        'flowEndMillisecondsTime',
        'flowDurationMilliseconds',
        'flowAttributes',
        'flowEndReason',
        'initialTCPFlags', 
        'unionTCPFlags',
        'reverseInitialTCPFlags',
        'packetTotalCount',
        'octetTotalCount',
        'reverseFlowAttributes',
        'reverseFlowDeltaMilliseconds',
        'reversePacketTotalCount',
        'reverseOctetTotalCount', 
        'smallPacketCount',
        'largePacketCount',
        'nonEmptyPacketCount',
        'firstNonEmptyPacketSize',
        'maxPacketSize',
        'dataByteCount',
        'bytesPerPacket',
        'averageInterarrivalTime',
        'standardDeviationInterarrivalTime',
        'reverseSmallPacketCount',
        'reverseLargePacketCount',
        'reverseNonEmptyPacketCount',
        'reverseFirstNonEmptyPacketSize',
        'reverseMaxPacketSize',
        'reverseDataByteCount',
        'reverseBytesPerPacket',
        'reverseAverageInterarrivalTime',
        'reverseStandardDeviationInterarrivalTime']]
```
