```python
import pandas as pd
import matplotlib.pyplot as plt
import math

df = pd.read_csv('/Users/lasyasrikantamneni/Downloads/mouse_computer_room_hub.csv') 


# Selecting only the numeric coloumns 
numeric_columns = df.select_dtypes(include=[float, int]).columns

num_columns = len(numeric_columns)
num_rows = math.ceil(num_columns / 5)

# Creating histogram 
df[numeric_columns].hist(bins=30, figsize=(15, 12), layout=(num_rows, 5))
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.suptitle('Mouse Computer Hub Histograms')
plt.show()
```
